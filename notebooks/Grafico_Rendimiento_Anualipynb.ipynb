```python
import pandas as pd
```

```python
df = pd.read_csv('/content/predicciones_Close_price.csv')
print(df)
```

```python
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
df = pd.DataFrame(df)
```

```python
df
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
```

```python
# Configurar el grfico
fig, ax1 = plt.subplots(figsize=(15, 10))

# Grfico del SP500 predictivo (lnea)
ax1.plot(df['Date '], df['Dolar_Close'], color='blue', marker='o', label='Dolar_Close')
ax1.set_xlabel('Date ')
ax1.set_ylabel('Dolar_Close', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Crear un segundo eje Y para el rendimiento anual
ax2 = ax1.twinx()
ax2.plot(df['Date '], df['Rendimiento _anual'], color='green', marker='x', linestyle='--', label='Rendimiento Anual (%)')
ax2.set_ylabel('Rendimiento _anual', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Ttulo y leyenda
plt.title('SP500 Predictivo y Rendimiento Anual')
fig.tight_layout()

# Mostrar el grfico
plt.show()
```
